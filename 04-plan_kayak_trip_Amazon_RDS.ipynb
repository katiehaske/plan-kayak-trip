Create and configure RDS instance on AWS. 

Use SQLAlchemy to create an engine that's connected to the remote database:

```python
# Change YOUR_USERNAME, YOUR_PASSWORD and YOUR_HOST with you credentials
engine = create_engine("postgresql+psycopg2://YOUR_USERNAME:YOUR_PASSWORD@YOUR_HOST/postgres", echo=True)
```

4. Create two tables, `cities` and `hotels`, in remote database with info saved in `.csv` files from previous sections.

```python
cities.to_sql("cities", engine)
hotels.to_sql("hotels", engine)
```

Use PGAdmin to explore the database

Use SQL syntax to make requests that will get the list of the 30 best places to go for vacay.

```python
conn = engine.connect()

statement = "SELECT city_name, main_weather, day_temperature, hotels.name, description, score, url \
             FROM cities \
             INNER JOIN hotels \
             ON cities.city_id = hotels.city_id \
             WHERE rank=1 AND score IS NOT NULL \
             ORDER BY score DESC \
             LIMIT 30"

best_places = pd.read_sql(statement, conn)
```
